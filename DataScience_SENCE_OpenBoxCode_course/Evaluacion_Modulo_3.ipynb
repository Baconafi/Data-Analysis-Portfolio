```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Crear el siguiente DataFrame indexado:

diccionario = {
    'calorias':[420,380,390,490,300],
    'tiempo':[60,40,75,55,45]
}

dias = ['L','M','X','J','V']

df = pd.DataFrame(diccionario, index=dias)

df
```

```python
# Calcular la media, mediana, y desviación típica de ambas columnas.
print("La media de calorias es 396")
print("La media de tiempo es 55")
print("La mediana de calorias es 390")
print("La mediana de tiempo es 55")
print("La desviación estándar de calorias es 68.77")
print("La desviación estándar de tiempo es 13.69\n")


display(df.describe())
display(df['calorias'].median())
display(df['tiempo'].median())
```

```python
# Añadir otra columna booleana al DataFrame para ver si se ha cumplido el reto de quemar más de 400 calorías por hora.
df['reto'] =  (df['calorias'] / (df['tiempo']/60)) > 400
df
```

```python
# Filtrar el DataFrame y devolver otro DataFrame con las filas pares que cumplan que el número de calorías es mayor de 400.
reset_df = df.reset_index()
reset_df = reset_df[(reset_df.index % 2 == 0) & (reset_df['reto'] == True)]
nuevo_df = reset_df.set_index('index')

nuevo_df
```

```python
# Crear a partir del DataFrame una serie con los porcentajes de días que se ha conseguido el reto y los que no.
serie = df['reto'].value_counts(normalize=True).apply(lambda x: x*100)
print("Ha conseguido el reto un 60% de los días y no ha conseguido el reto un 40% de los días")
display(serie)
```

```python
# Crear un gráfico como el de más abajo que muestre la progresión de las calorías y tiempo durante la semana.
fig, ax = plt.subplots()
ax.plot(df.index, df['calorias'], marker="|", markersize=6, markeredgewidth=1.5)
ax.plot(df.index, df['tiempo'], marker=".", markersize=10)
ax.legend(['Calorias','Tiempo'])
plt.title('Evolución diaria')
plt.show()
```
