```python
import pandas as pd
import datetime
import pytz

# Obtener la zona horaria de Santiago de Chile (GMT-4 en verano, GMT-3 en invierno)
zona_horaria_santiago = pytz.timezone('Chile/Continental')

# Obtener la fecha y hora actual en la zona horaria de Santiago de Chile
fecha_hora_actual = datetime.datetime.now(zona_horaria_santiago)

# Diccionarios vacios para la bodega y facturas
productos_dic = {
    "Nombre_producto":[],
    "Precio_producto":[]
}

factura_dic = {
    "Nombre_cliente":[],
    "Fecha":[],
    "Nombre_producto":[],
    "Precio_producto":[]
}

# Lista de opciones para los menú
lista_menu = [1,2,3,4,5,6]
lista_sub = [1,2,3,4]
lista_sub_2 = [1,2,3,4]

# DataFrames de los diccionarios de bodega y facturas
productos_df = pd.DataFrame(productos_dic)
factura_df = pd.DataFrame(factura_dic)

# Funciones para las distintas opciones del menú
def agregar_productos():
  productos_dic['Nombre_producto'].append(nombre_producto)
  productos_dic['Precio_producto'].append(precio_producto)
  global productos_df
  productos_df = pd.DataFrame(productos_dic)

def agregar_facturas():
  factura_dic['Nombre_cliente'].append(nombre_cliente)
  factura_dic['Nombre_producto'].append(nombre_producto_factura)
  factura_dic['Precio_producto'].append(precio_producto_factura)
  factura_dic['Fecha'].append(fecha_hora_actual)
  global factura_df
  factura_df = pd.DataFrame(factura_dic)

def ver_resumen():
  cantidad = factura_df['Nombre_producto'].value_counts().sum()
  monto_total = factura_df['Precio_producto'].sum()
  return "Se han vendido", cantidad, "productos y el monto total vendido es", monto_total

def guardar_resumen(archivo=None):
  cantidad = factura_df['Nombre_producto'].value_counts().sum()
  monto_total = factura_df['Precio_producto'].sum()
  guardar_resumen = {archivo: [cantidad, monto_total]}
  return guardar_resumen

def eliminar_productos(borrar=None):
  global productos_df
  lista_vacia = []
  lista_vacia.append(borrar)
  mask = productos_df.isin(lista_vacia).any(axis=1)
  productos_df = productos_df.loc[~mask]

def eliminar_factura(borrar_2=None):
  global factura_df
  lista_vacia_2 = []
  lista_vacia_2.append(borrar_2)
  mask_2 = factura_df.isin(lista_vacia_2).any(axis=1)
  factura_df = factura_df.loc[~mask_2]

def eliminar_todo():
  global productos_df
  global factura_df
  productos_df = productos_df.drop(productos_df.index, inplace=True)
  factura_df = factura_df.drop(factura_df.index, inplace=True)

# Creacion del menu para el usuario y submenús
def mostrar_menu():
    print("\nMenú Facturación y Bodegas")
    print("1. Ingresar producto")
    print("2. Ingresar factura")
    print("3. Ver resumen")
    print("4. Ver detalle")
    print("5. Borrar datos")
    print("6. Salir")

def submenu_1():
  print("\n1. Nuevo producto")
  print("2. Guardar producto")
  print("3. Volver")

def submenu_2():
  print("\n1. Ingresar producto")
  print("2. Guardar")
  print("3. Salir")

def submenu_3():
  print("\n¿Desea guardar su resumen?")

def submenu_4():
  print("\n1. Ver ventas por usuario")
  print("2. Ver ventas por producto")
  print("3. Ver ventas diarias")
  print("4. Guardar vista")

def submenu_5():
  print("\n1. Elegir producto")
  print("2. Elegir factura")
  print("3. Eliminar todo")
  print("4. Volver")

while True:
  try:
    mostrar_menu()
    opcion = int(input("Seleccione una opción (número del 1 al 6): "))
    if opcion == 1 & opcion in lista_menu:
      nombre_producto = input("Ingrese nombre del producto: ")
      precio_producto = float(input("Ingrese precio del producto: "))
      submenu_1()
      opcion_sub = int(input("Seleccione una opción (número del 1 al 3): "))
      if opcion_sub == 1 and opcion_sub in lista_sub:
        agregar_productos()
      elif opcion_sub == 2 & opcion_sub in lista_sub:
        agregar_productos()
      elif opcion_sub == 3 & opcion_sub in lista_sub:
        print("\nHa retornado al menú principal")
      else:
        print("Por favor, seleccione una opción válida, intente de nuevo")
    elif opcion == 2 & opcion in lista_menu:
      nombre_cliente = input("Ingrese nombre del cliente: ")
      submenu_2()
      opcion_sub_2 = int(input("Seleccione una opción (número del 1 al 3): "))
      if opcion_sub_2 == 1 & opcion_sub_2 in lista_sub:
        nombre_producto_factura = input("Ingrese nombre del producto: ")
        precio_producto_factura = float(input("Ingrese precio del producto: "))
        agregar_facturas()
      elif opcion_sub_2 == 2 & opcion_sub_2 in lista_sub:
        nombre_producto_factura = input("Ingrese nombre del producto: ")
        precio_producto_factura = float(input("Ingrese precio del producto: "))
        agregar_facturas()
      elif opcion_sub_2 == 3 & opcion_sub_2 in lista_sub:
        print("\nOpciones finalizadas. Ha salido del menú")
        break
      else:
        print("\nPor favor, seleccione una opción válida, intente de nuevo")
    elif opcion == 3 & opcion in lista_menu:
      print(ver_resumen())
      submenu_3()
      opcion_sub_3 = input("Ingrese SI o NO ")
      if opcion_sub_3.upper() == "SI":
        archivo = input("Ingrese nombre del archivo: ")
        print(guardar_resumen(archivo))
      elif opcion_sub_3.upper() == "NO":
        print("\nHa retornado al menú principal")
      else:
        print("\nPor favor, seleccione una opción válida, intente de nuevo")
    elif opcion == 4 & opcion in lista_menu:
      submenu_4()
      opcion_sub_4 = int(input("Ingrese una opción: "))
      if opcion_sub_4 == 1 & opcion_sub_4 in lista_sub_2:
        print(factura_df['Nombre_cliente'])
        usuario = input("Ingrese nombre de usuario ")
        print(factura_df[factura_df['Nombre_cliente']==usuario])
      elif opcion_sub_4 == 2 & opcion_sub_4 in lista_sub_2:
        print(factura_df['Nombre_producto'])
        prod = input("Ingrese producto ")
        print(factura_df[factura_df['Nombre_producto']==prod])
      elif opcion_sub_4 == 3 & opcion_sub_4 in lista_sub_2:
        factura_df['Fecha'] = pd.to_datetime(factura_df['Fecha'])
        factura_df['Dia'] = factura_df['Fecha'].dt.day
        print(factura_df.groupby('Dia')['Precio_producto'].sum())
      elif opcion_sub_4 == 4 & opcion_sub_4 in lista_sub_2:
        print(factura_df)
      else:
        print("\nPor favor, seleccione una opción válida, intente de nuevo")
    elif opcion == 5 & opcion in lista_menu:
      submenu_5()
      opcion_sub_5 = int(input("Ingrese una opción: "))
      if opcion_sub_5 == 1 & opcion_sub_5 in lista_sub_2:
        while True:
          print("\nPuede elegir los siguientes productos a eliminar: ", productos_df['Nombre_producto'].tolist())
          borrar = input("Elegir productos a eliminar. Para detenerse ingrese F: ")
          eliminar_productos(borrar)
          if borrar.upper() == 'F':
            break
      elif opcion_sub_5 == 2 & opcion_sub_5 in lista_sub_2:
        while True:
          print("\nPuede elegir las siguientes facturas a eliminar: ", factura_df['Nombre_producto'].tolist())
          borrar_2 = input("Elegir productos a eliminar. Para detenerse ingrese F: ")
          eliminar_factura(borrar_2)
          if borrar_2.upper() == 'F':
            break
      elif opcion_sub_5 == 3 & opcion_sub_5 in lista_sub_2:
        eliminar_todo()
        print("\nSe han eliminado todos los datos")
      elif opcion_sub_5 == 4 & opcion_sub_5 in lista_sub_2:
        print("\nHa retornado al menú principal")
      else:
        print("Por favor, seleccione una opción válida, intente de nuevo")
    elif opcion == 6 & opcion in lista_menu:
      print("\nOpciones finalizadas. Ha salido del menú")
      break
    else:
      print("\nPor favor, seleccione una opción válida, intente de nuevo")
  except ValueError:
    print("Error: Debe ingresar un número entero")
  except KeyError:
    print("Error: Clave no encontrada en el diccionario")

```
